### Attempt 2: Read data, fill in missing values by:

__1) aggregate mean at level of granularity of 'Country'__

__2) if NaN values still exist, replace with mean of whole column__

```python
# read data again
life_data = pd.read_csv('Life Expectancy Data.csv')
life_data 
```

```python
# Rename some column names as they contain trailing spaces.
life_data.rename(
    columns={" BMI ":"BMI","Life expectancy ":"Life_Expectancy","Adult Mortality":"Adult_Mortality",
             "infant deaths":"Infant_Deaths","percentage expenditure":"Percentage_Exp","Hepatitis B":"HepatitisB",
             "Measles ":"Measles"," BMI ":"BMI","under-five deaths ":"Under_Five_Deaths","Diphtheria ":"Diphtheria",
             " HIV/AIDS":"HIV/AIDS"," thinness  1-19 years":"thin_1_19_y"," thinness 5-9 years":"thin_5_9_y",
             "Income composition of resources":"Income_Comp","Total expenditure":"Tot_Exp"},
    inplace=True)
```

```python
life_data.head()
```

```python
life_data.isna().sum()
```

```python
# Columns with missing values
cols = ['Life_Expectancy', 'Adult_Mortality', 'Alcohol', 'HepatitisB',
       'BMI', 'Polio', 'Tot_Exp', 'Diphtheria', 'GDP', 'Population',
       'thin_1_19_y', 'thin_5_9_y', 'Income_Comp', 'Schooling']
```

```python
# fill in missing values with aggregate mean at level of granularity of 'Country'
for column in cols:
    life_data[column] = life_data.groupby('Country')[column].transform(lambda x: x.fillna(x.mean()))
```

```python
# same as above
for col in cols:
    life_data[col] = life_data[col].fillna(life_data.groupby('Country')[col].transform('mean'))
```

```python
life_data.isna().sum()
```

```python
# replace NaN values with average one column at a time using pandas
life_data['Life_Expectancy'].fillna((life_data['Life_Expectancy'].mean()), inplace=True)
life_data['Adult_Mortality'].fillna((life_data['Adult_Mortality'].mean()), inplace=True)
life_data['Alcohol'].fillna((life_data['Alcohol'].mean()), inplace=True)
life_data['HepatitisB'].fillna((life_data['HepatitisB'].mean()), inplace=True)
life_data['BMI'].fillna((life_data['BMI'].mean()), inplace=True)
life_data['Polio'].fillna((life_data['Polio'].mean()), inplace=True)
life_data['Tot_Exp'].fillna((life_data['Tot_Exp'].mean()), inplace=True)
life_data['Diphtheria'].fillna((life_data['Diphtheria'].mean()), inplace=True)
life_data['GDP'].fillna((life_data['GDP'].mean()), inplace=True)
life_data['Population'].fillna((life_data['Population'].mean()), inplace=True)
life_data['thin_1_19_y'].fillna((life_data['thin_1_19_y'].mean()), inplace=True)
life_data['thin_5_9_y'].fillna((life_data['thin_5_9_y'].mean()), inplace=True)
life_data['Income_Comp'].fillna((life_data['Income_Comp'].mean()), inplace=True)
life_data['Schooling'].fillna((life_data['Schooling'].mean()), inplace=True)
```

```python
# same as above
life_data['Life_Expectancy'] = life_data['Life_Expectancy'].fillna((life_data['Life_Expectancy'].mean()))
life_data['Adult_Mortality'] = life_data['Adult_Mortality'].fillna((life_data['Adult_Mortality'].mean()))
life_data['Alcohol'] = life_data['Alcohol'].fillna((life_data['Alcohol'].mean()))
life_data['HepatitisB'] = life_data['HepatitisB'].fillna((life_data['HepatitisB'].mean()))
life_data['BMI'] = life_data['BMI'].fillna((life_data['BMI'].mean()))
life_data['Polio'] = life_data['Polio'].fillna((life_data['Polio'].mean()))
life_data['Tot_Exp'] = life_data['Tot_Exp'].fillna((life_data['Tot_Exp'].mean()))
life_data['Diphtheria'] = life_data['Diphtheria'].fillna((life_data['Diphtheria'].mean()))
life_data['GDP'] = life_data['GDP'].fillna((life_data['GDP'].mean()))
life_data['Population'] = life_data['Population'].fillna((life_data['Population'].mean()))
life_data['thin_1_19_y'] = life_data['thin_1_19_y'].fillna((life_data['thin_1_19_y'].mean()))
life_data['thin_5_9_y'] = life_data['thin_5_9_y'].fillna((life_data['thin_5_9_y'].mean()))
life_data['Income_Comp'] = life_data['Income_Comp'].fillna((life_data['Income_Comp'].mean()))
life_data['Schooling'] = life_data['Schooling'].fillna((life_data['Schooling'].mean()))
```
